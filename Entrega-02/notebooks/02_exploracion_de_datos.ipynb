```python
data_train = np.load("/tmp/brain_tumor_images_training.npz")
```

```python
#data_train["x"].shape
columnas = 3
filas = 3
indices = np.random.randint(0,5739,(columnas*filas))
imgs = data_train["x"][indices.tolist()]
clases = data_train["y"][indices.tolist()]
```

```python


fig = plt.figure(figsize=(5*columnas, 5*filas))

ax = []

for i in range(columnas*filas):
    img = imgs[i]
    ax.append( fig.add_subplot(filas, columnas, i+1) )
    ax[-1].set_title("(" + str(indices[i]) + ")" + str(clases[i]))  # set title
    ax[-1].axis("off")
    plt.imshow(img, alpha=0.9)

plt.show()
```
