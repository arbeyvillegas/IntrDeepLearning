```python
data_test = np.load("/tmp/brain_tumor_images_test.npz")
etiquetas = ['glioma_tumor','no_tumor','meningioma_tumor','pituitary_tumor']
```

# **Preparar los datos para predicción**

```python
X_test = data_test["x"].astype('float32')
X_test /= 255
Y_test = np.array([etiquetas.index(item) for item in data_test['y']])
```

```python
#precision    recall  f1-score   support
y_pred = model.predict(X_test, batch_size=64, verbose=1)
y_pred_bool = np.argmax(y_pred, axis=1)
print(classification_report(Y_test, y_pred_bool))
print()
print()
```

```python
carpeta_training = "/tmp/brain_tumor_data/Training"
carpeta_testing = "/tmp/brain_tumor_data/Testing"
```

```python
from keras.preprocessing.image import ImageDataGenerator
batch_size = 32
nb_classes = 4
nb_epochs = 10
img_channel = 3
nb_filters = 32
nb_pool = 2
nb_conv = 3
img_size = 224

test_datagen = ImageDataGenerator(rescale=1./255)
validation_generator = test_datagen.flow_from_directory(carpeta_testing, target_size=(img_size, img_size), batch_size=batch_size, class_mode='categorical')
```

```python
!wget --no-check-certificate \
    "https://github.com/arbeyvillegas/IntrDeepLearning/raw/main/Entrega-02/mod_entr/VGG16model94accuracy_0.29-accuracy.keras" \
    -O "/tmp/modelo-transfer.keras"
```

```python
ruta_arch_model = "/tmp/modelo-transfer.keras"

model = tf.keras.models.load_model(ruta_arch_model, compile=True)


y_pred = model.predict(validation_generator, verbose=1)
#print('Test loss:', scores[0])
#print('Test accuracy:', scores[1])
```

```python
predicted_labels = np.argmax(y_pred, axis=1)
true_labels = validation_generator.classes
```

```python
print(classification_report(true_labels, predicted_labels))
```
